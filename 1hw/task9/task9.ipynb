```python
def func(str):
    str=str.lower().replace(" ","")
    for i in range (len(str)//2):
        if str[i]!=str[len(str)-i-1]:
            print("Строка не полиндром")
            return False

    print ("Строка является полиндромом")
    return True
```

```python
str=input("Введите строку: ")
result=func(str)
```
