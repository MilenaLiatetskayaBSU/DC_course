```python
import random
```

```python
def mean (numbers):
    result = 0
    
    for numb in numbers:
        result += numb
        
    return result/len(numbers)
```

```python
numbers = random.sample(range(1,101),5)
res = mean(numbers)

print("Исходный массив чисел: ", numbers, "\nСреднее арифметическое: ", res)
```
